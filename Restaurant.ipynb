```python
class restaurant:
    rname='Delicious'
    address='dilshuknagar'
    contact=7894561230
    customers_details={}
    def __init__(self,name,phone,password):
            self.name=name
            self.phone=phone 
            self.password=password
            self.concat_customer_details(self.phone,self) 

    @classmethod
    def enter(cls):
        n=int(input("select an option\n1.Register\n2.Login"))
        match n:
            case 1:
                phone=int(input("enter your phone number"))
                if phone in cls.customers_details:
                    print("account already registered\nif you want to continue to login press 1")
                    n=int(input("press 1 to login"))
                    if n==1:
                        phone=phone
                        password=input("enter your password")
                        cls.login(phone,password)
                    else:
                        cls.enter(cls)
                        
                else:
                    name=input("enter your name")
                    password=input("create your password")
                    cls.register(name,phone,password)
            case 2:
                phone=int(input("enter your phone number"))
                password=input("enter your password")
                cls.login(phone,password)
    
    @classmethod
    def concat_customer_details(cls,phone,addr):
        if phone in cls.customers_details:
            print("account already registered\nif you want to continue to login press 1")
            n=int(input("press 1 to login"))
            if n==1:
                phone=phone
                password=input("enter your password")
                cls.login(phone,password)
            else:
                return
        else:
            cls.customers_details[phone]=addr

    @classmethod
    def register(cls,name,phone,password):
        name=restaurant(name,phone,password)
        cls.login(phone,password)

    @classmethod
    def login(cls,phone,password):
        if phone in cls.customers_details and password==cls.customers_details[phone].password:
            print("You are in restaurant page".center(150,'.'))
            account.menu()
        else:
            print("enter correct credentials")
```

```python
import datetime
class account:
    starter_items={1:["Veg Manchuria",60],2:['Chicken Manchuria',180],3:['Spring Rolls',75],4:['Aloo 65',160],5:['Chilli Chicken',140],6:['Paneer 65',169],7:['Chilli Paneer',129]}
    maincourse_items={1:["Paneer Biryani",250],2:["Hyderabad Chicken DUM Biryani",200],3:["Hyderabad Mutton DUM Biryani",300],4:["White Rice and Prawns Curry ",350],5:["Curd Rice",160]}
    dessert_items={1:['Choco Lava Cake',109],2:['Chocolate Truffle Pastry',199],3:['Hazenut Brownie',199],4:['Pulla ice',19]}
    beverages_items={1:['ThumpsUp',100],2:['Coca Cola',110],3:['Blueberry Mojito',104],4:['Coffee',40]}
    cart={}
    confirm=[]
    orders_history={}
    order_count=0
    z=0
    
    count=1
    @classmethod
    def menu(cls,j=0):
        if j==0:
            cls.cart={}
            cls.confirm=[]
            
        n=int(input("select an option\n1.STARTERS\n2.MAIN COURSE\n3.DESSERTS\n4.BEVERAGES"))      
        match n:
            case 1:
                print("STARTERS_LIST".center(150,'-'))
                cls.starter()

            case 2:
                print("MAINCOURSE_LIST".center(150,'-'))
                cls.maincourse()
            case 3:
                print("DESSERT_LIST".center(150,'-'))
                cls.dessert()
            case 4:
                print("BEVERAGES_LIST".center(150,'-'))
                cls.beverages()
    
    @classmethod
    def starter(cls,j=0):
        if j==0:
            for i in cls.starter_items:
                print(i,cls.starter_items[i][0],'RS:',cls.starter_items[i][1])
            print(0,"Main Menu")    
            print(420,'checkout')
            
           
        else:
            n=int(input("select an item\ns1\n2\n3\n4\n5\n6\n7\n0\n420"))
            if 1<=n<=len(cls.starter_items):
                cls.cart[cls.count]=cls.starter_items[n]
                cls.count+=1
            elif n==420:
                print('Your_CartList'.center(150,'-'))
                cls.checkout()
                return
            else:
                cls.menu(j+1)
        if cls.z==0:
            return cls.starter(j+1)

    @classmethod
    def maincourse(cls,j=0):
        if j==0:
            for i in cls.maincourse_items:
                print(i,cls.maincourse_items[i][0],'RS:',cls.maincourse_items[i][1])
            print(0,"Main Menu")    
            print(420,'checkout')
        else:
            n=int(input("select an item\nm1\n2\n3\n4\n5\n0\n420"))
            if 1<=n<=len(cls.maincourse_items):
                cls.cart[cls.count]=cls.maincourse_items[n]
                cls.count+=1
            elif n==420:
                print('Your_CartList'.center(150,'-'))
                cls.checkout()
                return
            else:
                cls.menu(j+1)
        if cls.z==0:
            return cls.maincourse(j+1)
    
    @classmethod
    def dessert(cls,j=0):
        if j==0:
            for i in cls.dessert_items:
                print(i,cls.dessert_items[i][0],'RS:',cls.dessert_items[i][1])
            print(0,"Main Menu")    
            print(420,'checkout')
        else:
            n=int(input("select an item\nd1\n2\n3\n4\n0\n420"))
            if 1<=n<=len(cls.dessert_items):
                cls.cart[cls.count]=cls.dessert_items[n]
                cls.count+=1
            elif n==420:
                print('Your_CartList'.center(150,'-'))
                cls.checkout()
                return
            else:
                cls.menu(j+1)
        if cls.z==0:
            return cls.dessert(j+1)
    
    @classmethod
    def beverages(cls,j=0):
        if j==0:
            for i in cls.beverages_items:
                print(i,cls.beverages_items[i][0],'RS:',cls.beverages_items[i][1])
            print(0,"Main Menu")    
            print(420,'checkout')
        else:
            n=int(input("select an item\nb1\n2\n3\n4\n0\n420"))
            if 1<=n<=len(cls.beverages_items):
                cls.cart[cls.count]=cls.beverages_items[n]
                cls.count+=1
            elif n==420:
                print('Your_CartList'.center(150,'-'))
                cls.checkout()
                return
            else:
                cls.menu(j+1)
        if cls.z==0:
            return cls.beverages(j+1)

    @classmethod
    def checkout(cls,j=0,temp=[]):
        if j==0:
            for i in cls.cart:
                print(i,cls.cart[i][0],cls.cart[i][1])
    
        n=int(input("select an item number to add"))
       
        if n>len(cls.cart):
            print("please enter item number present in the cart")
                
            return cls.checkout(j+1)
        
        else:
            m=int(input("select its quantity"))
            if len(cls.confirm)!=0:
                for k in range(len(cls.confirm)):
                    if cls.cart[n][0]==cls.confirm[k][0]:
                        cls.confirm[k][1]+=m
                        cls.confirm[k][2]+=cls.cart[n][1]*m
                        break
                else:
                    cls.confirm+=[[cls.cart[n][0],m,cls.cart[n][1]*m]]
            else:
                cls.confirm+=[[cls.cart[n][0],m,cls.cart[n][1]*m]]
        x=int(input("select an option\n1.Proceed to Payment\n2.Add more items from Cart"))
        if x==1:
            cls.temp=cls.confirm
            # cls.cart={}
            # cls.confirm={}
            cls.count=1
            
            print("Bill_Reciept".center(150,'.'))
            cls.bill_generate()
            return
        else:
            cls.checkout(j+1)


    @classmethod
    def bill_generate(cls,j=1,total=0):
        print('s.no'.center(35),'item_name'.center(35),'quantity'.center(35),'amount'.center(35),sep=' | ')
        for i in cls.temp:
            
            print(str(j).center(35),'| ',i[0].center(34),'| ',str(i[1]).center(35),'| ',str(i[2]).center(35))
            total+=i[2]
            j+=1
        print(('_'*30).center(35),('_'*30).center(190))
        print("Total Amount:".center(35),'=',str(total).center(190))
        print(".".center(149,'.'))
        cls.payment()
    @classmethod
    def payment(cls):
        n=int(input("select an option\n1.pay\n2.exit"))
        if n==1:
            
            print("ORDER PLACED......")
            print("Your Order is on the way...")
            cls.concat_order()
            cls.z=1
        else:
            return
        

        
    @classmethod
    def concat_order(cls):
        cls.orders_history[cls.order_count+1]=cls.temp,datetime.datetime.now()
        cls.order_count+=1


    @classmethod
    def orders(cls):
        print("Total Number of Orders:",cls.order_count,'-'*150)
        print("s.no".center(25,'.'), "Order Details".center(130,'.'),sep=' | ')
        for i in cls.orders_history:
            print(str(i).center(25),cls.orders_history[i],sep=' | ')
                


    
```
